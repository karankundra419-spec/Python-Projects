Task 2: Compute missing Project Cost values using running average (use for-loop)
==

```python
# Example: project_df already exists
# project_df["Cost"] = pd.to_numeric(project_df["Cost"], errors="coerce")

project_df["Cost"]= pd.to_numeric(project_df["Cost"], errors="coerce")
project_df
```

```python
# Dataframe
Project = pd.DataFrame({
    'ID': ["A001", "A002", "A003", "A004", "A005", "A002", "A005", "A003", "A001", "A003", "A001", "A004", "A004", "A005"],
    "Project":["Project 1", "Project 2", "Project 3", "Project 4", "Project 5", "Project 6", "Project 7", "Project 8", "Project 9", "Project 10","Project 11", "Project 12", "Project 13", "Project 14"],
    "Cost": [1002000, 2000000, 4500000, 5500000, np.nan, 680000, 400000, 350000, np.nan, 300000, 2000000, 1000000, 3000000, 200000],
    "Status": ["Finished", "Ongoing", "Finished", "Ongoing", "Finished", "Failed", "Finished", "Failed", "Ongoing", "Finished", "Failed", "Ongoing", "Finished", "Finished"]
    
})

print("Original DataFrame:")
print(Project)

# Replace missing values in 'cost' with running average using a for loop
for i in range(len(Project)):
    if pd.isna(Project.loc[i, "Cost"]):  # check if value is NaN
        # Take average of all previous non-missing values
        prev_values = Project.loc[:i-1, "Cost"].dropna()
        if len(prev_values) > 0:
            Project.loc[i, "Cost"] = prev_values.mean()

print("\nDataFrame after replacing missing values:")
print(Project)

Project.to_csv("projects.csv")
```

Task 3: Split Name into First Name and Last Name and drop old Name column

```python
employee = pd.DataFrame({   
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],  # employee IDs  
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],  # full names  
    'Gender': ['M', 'F', 'M', 'F', 'F'],  # genders  
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],  # cities  
    'Age': [25, 27, 29, 31, 30]})  # ages  
print("Original DataFrame:")
print(employee)


employee[['First_name', 'Last_name']]= employee['Name'].str.split(' ',n=1, expand=True)

employee.drop(columns=['Name'], inplace=True)

print("\nDataFrame after Split name into first_name and Last_Name:")
print(employee)

employee.to_csv("employees.csv")
```

Task 4: Join all three DataFrames into a single DataFrame 
===

```python
employee_df = pd.read_csv("employees.csv")
project_df = pd.read_csv("projects.csv")
seniority_df = pd.read_csv("seniority.csv")
print(employee_df)
print(project_df)
print(seniority_df)
```

```python
merged_df = employee_df.merge(seniority_df, on="ID", how='left')
final = merged_df.merge(project_df, on="ID", how='left')
final.to_csv('Final.csv', index=False)
```

Task 5: Add a 5% bonus for employees who finished projects
==

```python
# Add Bonus column equal to 5% of Project Cost when Status is 'Finished', else 0  # comment
final['Bonus'] = np.where(final['Status'].str.strip().str.lower() == 'finished', final['Cost'] * 0.05, 0.0)  # compute bonus per row  
# Save after adding bonus for record-keeping  # comment
final.to_csv('Final_with_bonus.csv', index=False)  # save final with bonus to CSV  
```

```python
Final=pd.read_csv('final_with_bonus.csv')
print(Final)
```

Task 6: Demote designation by 1 if project status is 'Failed' and delete employees with designation >4
====

```python
failed_mask = final['Status'].str.strip().str.lower() == 'Failed'  # boolean mask for failed projects  
final.loc[failed_mask, 'Designation Level'] = final.loc[failed_mask, 'Designation Level'] - 1  # decrement designation level  
# Remove rows (employee records) where designation level is now above 4 (ineligible)  # comment
final = final[final['Designation Level'] <= 4].copy()  # keep only rows with designation <= 4  
# Save after demotion and deletion  # comment
final.to_csv('Final_after_demotion.csv', index=False)  # save file  
```

```python
final_status=pd.read_csv("final_after_demotion.csv")
print(final_status)
```

Task 7: Prefix First Name with Mr./Mrs. and drop Gender column
==

```python
# Add Mr. or Mrs. based on Gender and then drop the Gender column  # comment
final['First_name'] = np.where(final['Gender'] == 'M', 'Mr. ' + final['First_name'], 'Mrs. ' + final['First_name'])  # prefix titles  
final.drop(columns=['Gender'], inplace=True)  # drop the Gender column as requested  
# Save after adding titles  # comment
final.to_csv('Final_with_titles.csv', index=False)  # save file  
```

```python
final_with_prefix=pd.read_csv("Final_with_titles.csv")
print(final_with_prefix)
```

Task 8: Promote designation by 1 for employees whose Age > 29
==

```python
# Increase designation level by 1 for employees older than 29 years using an IF-style condition  # comment
age_mask = final['Age'] > 29  # boolean mask where age > 29  
final.loc[age_mask, 'Designation Level'] = final.loc[age_mask, 'Designation Level'] + 1  # increment designation level  
# Save after promotion step  # comment
final.to_csv('Final_after_promotion.csv', index=False)  # save file  
```

```python
final_after_promotion=pd.read_csv("Final_after_promotion.csv")
print(final_after_promotion)
```

Task 9: Aggregate total project cost per employee into TotalProjCost DataFrame
==

```python
# Group by ID and First Name to sum all Project Costs and create a new DataFrame with the results  
total_proj_cost = final.groupby(['ID', 'First_name'], as_index=False)['Cost'].sum()  # group and sum costs  
# Rename the aggregated column to 'Total cost' to match the brief  # comment
total_proj_cost.rename(columns={'Cost': 'Total cost'}, inplace=True)  # rename summed column  
# Save the TotalProjCost DataFrame to CSV as requested  # comment
total_proj_cost.to_csv('TotalProjCost.csv', index=False)  # save aggregated totals to CSV  
```
